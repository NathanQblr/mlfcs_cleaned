```python
from umap.umap_ import UMAP


def plot_umap(data_beads):
    """Plot data from classification in a UMAP 2D representation when we have 4 candidates models

    Args:
        data_beads: pd.DataFrame with prediction of models

    Returns:
        plt.figure
    """
    # Load predicted probabilities for each model
    prob_data = data_beads[['Proba MB', 'Proba FBM', 'Proba CTRW', 'Proba RWF']]

    # Apply UMAP for dimensionality reduction
    umap_model = UMAP()#(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
    embedding = umap_model.fit_transform(prob_data)

    # Add UMAP results to the DataFrame
    data_beads['UMAP1'] = embedding[:, 0]
    data_beads['UMAP2'] = embedding[:, 1]

    glycerol = np.unique(data_beads['Glycerol'])

    # Normalize 'file name' for color coding
    for gly in glycerol:
        name_encoder = LabelEncoder()
        name_encoder.fit(data_beads.loc[(data_beads['Glycerol'] == gly), 'file name'])
        data_beads.loc[(data_beads['Glycerol'] == gly), 'file name'] = name_encoder.transform(
            data_beads.loc[(data_beads['Glycerol'] == gly), 'file name'])

    # Create the UMAP plot
    cm = 1 / 2.54  # centimeters in inches
    sns.set_style("whitegrid")
    plt.rcParams['font.size'] = 8

    fig, axs = plt.subplots(2, int(np.ceil(len(glycerol) / 2)), figsize=(15 * cm, 15 * cm))

    k = 0
    for line in axs:
        for col in line:
            if k < len(glycerol):
                sns.scatterplot(
                    data=data_beads.loc[(data_beads['Glycerol'] == glycerol[k])],
                    x="UMAP1", y="UMAP2", hue="file name", palette='dark',
                    ax=col, legend=False, marker='o', s=15
                )
                col.set_aspect('equal', 'box')
                col.set(title=glycerol[k], frame_on=True)
                k += 1

    fig.tight_layout()
    return fig
```

```python
from mpl_toolkits.mplot3d import Axes3D


def plot_quaternary(data_beads):
    """Plot data from classification in a 3D quaternary plot for 4 candidate models.

    Args:
        data_beads: pd.DataFrame with prediction of models.

    Returns:
        plt.figure
    """
    # Load predicted probabilities for each model
    prob_data = data_beads[['Proba MB', 'Proba FBM', 'Proba CTRW', 'Proba RWF']]

    # Normalize the probabilities to ensure they sum to 1 for each point
    prob_data = prob_data.div(prob_data.sum(axis=1), axis=0)

    # Convert to Cartesian coordinates
    # A 3D quaternary plot (tetrahedron) has vertices at:
    # (1, 0, 0), (0, 1, 0), (0, 0, 1), (1/3, 1/3, 1/3) for the fourth class in 3D space
    vertices = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1], [1/3, 1/3, 1/3]])

    # Convert probabilities to 3D points
    coords = prob_data.dot(vertices)

    # Create a 3D plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the points
    scatter = ax.scatter(
        coords.iloc[:, 0], coords.iloc[:, 1], coords.iloc[:, 2],
        c=data_beads['file name'], cmap='viridis', s=15
    )

    # Set the labels
    ax.set_xlabel('X (Proba MB)')
    ax.set_ylabel('Y (Proba FBM)')
    ax.set_zlabel('Z (Proba CTRW)')

    # Add a color bar
    cbar = plt.colorbar(scatter, ax=ax, shrink=0.5, aspect=10)
    cbar.set_label('File Name (encoded)')

    # Set equal aspect ratio
    ax.set_box_aspect([1, 1, 1])  # Aspect ratio is 1:1:1

    # Plot tetrahedron edges for reference
    for i in range(len(vertices)):
        for j in range(i + 1, len(vertices)):
            ax.plot(
                [vertices[i, 0], vertices[j, 0]],
                [vertices[i, 1], vertices[j, 1]],
                [vertices[i, 2], vertices[j, 2]],
                color='black', linewidth=1
            )

    ax.set_title('3D Quaternary Plot of Model Probabilities')
    plt.tight_layout()

    return fig

```

```python


def plot_polar(data_beads):
    """Plot data from classification in a polar representation for 4 candidate models.

    Args:
        data_beads: pd.DataFrame with prediction of models.

    Returns:
        plt.figure
    """
    # Normalize the probabilities for each model
    prob_data = data_beads[['Proba MB', 'Proba FBM', 'Proba CTRW', 'Proba RWF']]
    prob_data = prob_data.div(prob_data.sum(axis=1), axis=0)  # Ensure they sum to 1

    # Calculate angles for each probability (dividing the circle into 4 equal parts)
    angles = np.linspace(0, 2 * np.pi, 5)[:-1]  # [0, π/2, π, 3π/2]
    labels = ['Proba MB', 'Proba FBM', 'Proba CTRW', 'Proba RWF']

    # Prepare polar coordinates
    data_beads['theta'] = prob_data.apply(lambda row: np.dot(row, angles), axis=1)
    data_beads['radius'] = prob_data.max(axis=1)  # Use the max probability as the radius

    glycerol = np.unique(data_beads['Glycerol'])

    # Normalize 'file name' for color coding
    for gly in glycerol:
        name_encoder = LabelEncoder()
        name_encoder.fit(data_beads.loc[(data_beads['Glycerol'] == gly), 'file name'])
        data_beads.loc[(data_beads['Glycerol'] == gly), 'file name'] = name_encoder.transform(
            data_beads.loc[(data_beads['Glycerol'] == gly), 'file name'])

    # Create the polar plot
    cm = 1 / 2.54  # centimeters in inches
    sns.set_style("whitegrid")
    plt.rcParams['font.size'] = 8

    fig, axs = plt.subplots(2, int(np.ceil(len(glycerol) / 2)), figsize=(15 * cm, 15 * cm), subplot_kw={'projection': 'polar'})

    k = 0
    for line in axs:
        for col in line:
            if k < len(glycerol):
                subset = data_beads.loc[data_beads['Glycerol'] == glycerol[k]]
                col.scatter(subset['theta'], subset['radius'], c=subset['file name'], cmap='viridis', s=15)
                col.set_theta_zero_location('N')  # Set 0 degrees to the top
                col.set_theta_direction(-1)  # Set the direction to clockwise
                col.set_title(glycerol[k], va='bottom')
                k += 1

    fig.tight_layout()
    return fig

```
